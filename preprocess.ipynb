```python
data = []
for file in os.listdir('./data/html'):
    if file.endswith('.html'):
        with open('./data//html/' + file, 'r') as f:
            html = f.read()
            tree = bs4.BeautifulSoup(html, 'html.parser')
            data.append((file, tree.get_text('\n')))
```

```python
df = pd.DataFrame(data, columns=['file', 'ori_text'])
df['text'] = df['ori_text'].str.replace(r'\n+', '#', regex=True)
df
```

```python
titles = set()
for text in df['text']:
    titles |= set(re.findall(r'【.+?】', text))
titles
```

```python
titles = re.compile(r'【(升级建议|散热分析|购买建议|猪王的良心结语|(CPU|显卡|处理器)(性能)?[简分]析|游戏测试)】')
```

```python
def parse_text(text):
    data = {}
    title = 'head'
    cursor = 0
    for match in re.finditer(titles, text):
        data[title] = text[cursor: match.start()]
        title, cursor = match.group(1), match.start()
        title = re.sub(r'(CPU|处理器)(性能)?[简分]析', '处理器分析', title)
    data[title] = text[cursor:]
    return data
```

```python
data = df.join(pd.DataFrame([parse_text(x) for x in df['text']]))
data['file'] = df['file'].values
data
```

```python
data['配置'] = data['head'].str.extract(r'它的配置如下：(.+)它的优缺点如下：')[0]
data['优缺点'] = data['head'].str.extract(r'它的优缺点如下：(.+)')[0]
data['优点'] = data['head'].str.extract(r'优点！(.+)缺点！')[0]
data['缺点'] = data['head'].str.extract(r'缺点！(.+)')[0]
data['head_short'] = data['head'].str.replace(r'它的配置如下：(.+)', '', regex=True)

data['屏幕方面'] = data['购买建议'].str.extract(r'(屏幕方面#.+)(?=#接口方面#|#噪音方面#|#续航方面#)')
data['接口方面'] = data['购买建议'].str.extract(r'(接口方面#.+)(?=#屏幕方面#|#噪音方面#|#续航方面#)')
temp = data['购买建议'].str.extract(r'(噪音方面#.+)(?=#接口方面#|#屏幕方面#|#续航方面#)|(噪音方面#.+?)#')
data['噪音方面'] = temp[0].fillna('') + temp[1].fillna('')
temp = data['购买建议'].str.extract(r'(续航方面#.+)(?=#接口方面#|#噪音方面#|#屏幕方面#)|(续航方面#.+?)#')
data['续航方面'] = temp[0].fillna('') + temp[1].fillna('')
```

```python
for i, row in data.iterrows():
    suggestion = row['购买建议']
    if pd.isna(suggestion):
        continue
    for label in ['屏幕方面', '接口方面', '噪音方面', '续航方面']:
        if pd.isna(row[label]):
            continue
        suggestion = suggestion.replace(row[label], '')
    data.at[i, 'suggestion_short'] = suggestion
```

```python
data = data.fillna('')
data['common'] = data[['head_short', '优缺点', 'suggestion_short', '猪王的良心结语']].agg('#'.join, axis=1)
data['common'] = data['common'].str.replace(r'[#\s]{2}', '#', regex=True)
```

```python
def get_propmt(feature):
    return f'你是一个信息抽取模型，你负责从一篇笔记本电脑的评测文章中帮我提取该产品的“{feature}”描述。如未找到相关描述则回复“未描述”。\n'
```

```python
with pd.ExcelWriter('./prompts2.xlsx') as writer:
    for feature, src, explain in sheets:
        temp = data[['file']].copy()
        temp['prompt'] = get_propmt(feature) + '#' + explain + '\ntext: ' + data[src] + f'\n请提取text中“{feature}”描述'
        temp['answer'] = pd.NA
        temp.to_excel(writer, sheet_name=feature, index=False)
```
